```python
def confusion_matrix(classes, test_targets, prediction):
    n = len(classes)
    matrix = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            matrix[i, j] = np.sum((test_targets == classes[i]) & (prediction == classes[j]))
    return matrix
```

```python
confusion_matrix(c, y, pred)
```

```python
def calculate_accuracy(y_true, y_pred):
    return sum(1 for true, pred in zip(y_true, y_pred) if true == pred) / len(y_true)
```

```python
calculate_accuracy(y, pred)
```
