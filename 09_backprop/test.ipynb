```python
X_test = test_features
W1 = W1tr
W2 = W2tr

guesses = test_nn(X_test, M, K, W1, W2)
print(guesses)
```

```python
accuracy = (guesses == test_targets).mean()
print("accuracy: ",accuracy)

def confusion_matrix(classes, test_targets, prediction):
    n = len(classes)
    matrix = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            matrix[i, j] = np.sum((test_targets == classes[i]) & (prediction == classes[j]))
    return matrix

# Print the confusion matrix
print("Confusion Matrix:")
print(confusion_matrix([0, 1, 2], test_targets, guesses))

plt.figure(figsize=(12, 6))
plt.plot(range(iterations), Etotal)
plt.title("E_total as a function of iterations")
plt.xlabel("Iterations")
plt.ylabel("E_total")
plt.grid()

# Plot misclassification_rate
plt.figure(figsize=(12, 6))
plt.plot(range(iterations), misclassification_rate)
plt.title("Misclassification Rate as a function of iterations")
plt.xlabel("Iterations")
plt.ylabel("Misclassification Rate")
plt.grid()
```
